```r
library(ggplot2)
library(ggpubr)
```

```r
df = data.frame()
for (ncell in c(2500,1000,100)){
    for (transf in c('clr', 'alr')){
        for (impute in c('None', 'CZM', 'GBM', 'lrSVD', 'freqshrink')){
            for (i in 1:200){
                f = paste0('../results_review/sampling_original/genes770/mse/mse_', ncell, '_500_', i, '_', impute, '_', transf, '.csv')
                tmp = read.csv(f)
                tmp['ncell'] = ncell
                tmp['transf'] = transf
                tmp['impute'] = impute
                df = rbind(df, tmp)
            }
        }
    }
}
```

```r
df$impute = factor(df$impute, levels=c('None','CZM','GBM','lrSVD','freqshrink'))
```

# Partial correlation

```r
# organize data
tmp             = df[df$corr == 'partial correlation',]
pos             = which(tmp$shrink == 'shrink std' & tmp$transf == 'clr')
tmp$shrink[pos] = 'shrink CLR'
pos             = which(tmp$shrink == 'shrink std' & tmp$transf == 'alr')
tmp$shrink[pos] = 'shrink ALR'
for (i in c('no shrinkage', 'shrink basis')){
    pos = which(tmp$shrink == i & tmp$transf == 'alr')
    tmp = tmp[-pos,]
}
tmp$shrink = factor(tmp$shrink, levels=c('no shrinkage', 'shrink CLR', 'shrink ALR', 'shrink basis'))

# N=2500
tmp0 = tmp[tmp$ncell == 2500,]
g0   = ggplot(tmp0, aes(x=shrink, y=mse, fill=impute)) + 
        geom_boxplot(outlier.shape = NA, notch=T, width=0.7) +
        scale_y_continuous(labels = function(x) formatC(x, format = 'e', digits=2)) +
        scale_x_discrete(labels=c('no\nshrinkage', 'shrink\nCLR', 'shrink\nALR', 'shrink\nbasis')) +
        xlab('') + 
        ylab('Mean squared error') +
        labs(fill='Imputation') +
        ggtitle('MSE partial correlation matrix, D=500, N=2500') +
        theme(
            plot.title=element_text(face="bold", size=18), 
            axis.text.x = element_text(size=18), 
            axis.text.y=element_text(size=18),
            axis.title.y=element_text(size=18),
            legend.text = element_text(size = 18),
            legend.title = element_text(size=18)
        )

# N=1000
tmp1 = tmp[tmp$ncell == 1000,]
g1   = ggplot(tmp1, aes(x=shrink, y=mse, fill=impute)) + 
        geom_boxplot(outlier.shape = NA, notch=T, width=0.7) +
        scale_y_continuous(labels = function(x) formatC(x, format = 'e', digits=2)) +
        scale_x_discrete(labels=c('no\nshrinkage', 'shrink\nCLR', 'shrink\nALR', 'shrink\nbasis')) +
        xlab('') + 
        ylab('Mean squared error') +
        labs(fill='Imputation') +
        ggtitle('MSE partial correlation matrix, D=500, N=1000') +
        theme(
            plot.title=element_text(face="bold", size=18), 
            axis.text.x = element_text(size=18), 
            axis.text.y=element_text(size=18),
            axis.title.y=element_text(size=18),
            legend.text = element_text(size = 18),
            legend.title = element_text(size=18)
        )
# options(repr.plot.width=14, repr.plot.height=6)           
g01 = ggarrange(g0, g1, widths=c(1,1), legend='none')
# g01                        

# N=100
tmp2  = tmp[tmp$ncell == 100,]   
tmp21 = tmp2[tmp2$shrink == 'no shrinkage',]
tmp22 = tmp2[tmp2$shrink != 'no shrinkage',]
g21   = ggplot(tmp21, aes(x=shrink, y=mse, fill=impute)) + 
        geom_boxplot(outlier.shape = NA, notch=T, width=0.52) +
        scale_y_continuous(labels = function(x) formatC(x, format = 'e', digits=2)) +
        xlab('') + 
        ylab('Mean squared error') +
        labs(fill='Imputation') +
        theme(
            axis.text.x = element_text(size=18), 
            axis.text.y=element_text(size=18),
            axis.title.y=element_text(size=18),
            legend.text = element_text(size = 18),
            legend.title = element_text(size=18)
        ) 
g22   = ggplot(tmp22, aes(x=shrink, y=mse, fill=impute)) + 
        geom_boxplot(outlier.shape = NA, notch=T, width=0.52) +
        scale_y_continuous(labels = function(x) formatC(x, format = 'e', digits=2)) +
        xlab('') + 
        ylab('') +
        labs(fill='Imputation') +
        theme(
            axis.text.x = element_text(size=18), 
            axis.text.y=element_text(size=18),
            axis.title.y=element_blank(),
            legend.text = element_text(size = 18),
            legend.title = element_text(size=18)
        ) 
# options(repr.plot.width=10, repr.plot.height=6.1) 
b =  1 / 2 - 0.015
a = b / 2
c = 1 - b - a
g2    = ggarrange(g21, g22, NULL, nrow=1, widths=c(a,b,c), legend='bottom', common.legend=T)       
g2    = annotate_figure(g2, top = text_grob('MSE partial correlation matrix, D=500, N=100', color = "black", face = "bold", size = 18))
# g2
                           
options(repr.plot.width=15, repr.plot.height=12)       
ggarrange(g01, g2, nrow=2)
```

# Covariance

```r
# organize data
tmp             = df[df$corr == 'covariance' & df$transf == 'clr',]
pos             = which(tmp$shrink == 'shrink std')
tmp$shrink[pos] = 'shrink CLR'
tmp$shrink      = factor(tmp$shrink, levels=c('no shrinkage', 'shrink CLR', 'shrink basis'))


# N=2500
tmp0 = tmp[tmp$ncell == 2500,]
g0_clr = ggplot(tmp0, aes(x=shrink, y=mse, fill=impute)) + 
        geom_boxplot(outlier.shape = NA, notch = TRUE) +
        scale_y_continuous(labels = function(x) format(x, scientific = TRUE)) +
        xlab('') + 
        ylab('Mean squared error') +
        labs(fill='Imputation') +
        ggtitle('MSE covariance matrix (CLR), D=500, N=2500') +
        theme(
            plot.title=element_text(face="bold", size=18), 
            axis.text.x = element_text(size=18), 
            axis.text.y = element_text(size=17),
            axis.title.y=element_text(size=18),
            legend.text = element_text(size = 18),
            legend.title = element_text(size=18)
        )
                           
# N=1000
tmp1 = tmp[tmp$ncell == 1000,]
g1_clr = ggplot(tmp1, aes(x=shrink, y=mse, fill=impute)) + 
        geom_boxplot(outlier.shape = NA, notch = TRUE) +
        scale_y_continuous(labels = function(x) format(x, scientific = TRUE)) + 
        xlab('') + 
        ylab('Mean squared error') +
        labs(fill='Imputation') +
        ggtitle('MSE covariance matrix (CLR), D=500, N=1000') +
        theme(
            plot.title=element_text(face="bold", size=18), 
            axis.text.x = element_text(size=18), 
            axis.text.y = element_text(size=17),
            axis.title.y=element_text(size=18),
            legend.text = element_text(size = 18),
            legend.title = element_text(size=18)
        )
                           
# N=100
tmp2 = tmp[tmp$ncell == 100,]   
tmp21 = tmp2[tmp2$shrink == 'no shrinkage',]
tmp22 = tmp2[tmp2$shrink != 'no shrinkage',]
g21_clr = ggplot(tmp21, aes(x=shrink, y=mse, fill=impute)) + 
        geom_boxplot(outlier.shape = NA, notch = TRUE) +
        scale_y_continuous(labels = function(x) format(x, scientific = TRUE)) + 
        xlab('') + 
        ylab('Mean squared error') +
        labs(fill='Imputation') +
        theme(
            plot.title=element_text(face="bold", size=18), 
            axis.text.x = element_text(size=18), 
            axis.text.y = element_text(size=17),
            axis.title.y=element_text(size=18),
            legend.text = element_text(size = 18),
            legend.title = element_text(size=18)
        )
g22_clr = ggplot(tmp22, aes(x=shrink, y=mse, fill=impute)) + 
        geom_boxplot(outlier.shape = NA, notch = TRUE) +
        scale_y_continuous(labels = function(x) format(x, scientific = TRUE)) + 
        xlab('') + 
        ylab('') +
        labs(fill='Imputation') +
        theme(
            plot.title=element_text(face="bold", size=18), 
            axis.text.x = element_text(size=18), 
            axis.text.y = element_text(size=17),
            axis.title.y=element_blank(),
            legend.text = element_text(size = 18),
            legend.title = element_text(size=18)
        )
g2_clr = ggarrange(g21_clr, g22_clr, widths=c(0.5,1), legend='none')       
g2_clr = annotate_figure(g2_clr, top = text_grob('MSE covariance matrix (CLR), D=500, N=100', color = "black", face = "bold", size = 18))

```
